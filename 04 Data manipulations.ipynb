```julia
using JuliaDB, OnlineStats, StatPlots # IndexedTables,
```

```julia
# http://sdwebx.worldbank.org/climateportal/index.cfm?page=downscaled_data_download&menu=historical

dt = loadtable("C:\\Users\\ITC\\Downloads\\weather_1901_2015.csv")
```

```julia
dt = renamecol(dt,:_Year, :Year)
dt = renamecol(dt,:tas, :Temperature)
dt = select(dt, (:Year, :Month, :Country, :Temperature))
```

```julia
arr = ["SAU", "ARE", "BHR", "EGY", "KWT", "OMN","QAT","TUR", "USA"]
ARR = ["SAUDI ARABIA", "EMIRATES", "BAHRAIN", "EGYPT", "KWAIT", "OMAN","QATAR","TURKY", "USA"]
for (index, country) in enumerate(arr)
       dt = setcol(dt, :Country, :Country=>x->(x==country ? ARR[index] : x))
end
```

```julia
dt
```

```julia
dts = summarize((minimum,maximum, mean),dt, (:Country, :Year), select=(:Temperature))
```

```julia
#gr(fmt = :png) # choose the fast GR backend and set format to png: svg would probably crash with so many points
plotly()
@df dts scatter(:Year, (:mean), group=:Country,layout = 1, legend = :bottomleft)
```
